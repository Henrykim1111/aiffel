## 6-2. 머신러닝 알고리즘
- 지도학습 (Supervised Learning)
- 비지도학습 (Unsupervised Learning)
- 강화학습 (Reinforcement Learning)

### 지도 학습과 비지도 학습
---
- 적절한 머신러닝 알고리즘을 선택하는 방법 
    (https://modulabs.co.kr/blog/choosing-right-machine-learning-algorithm/)

1. 지도 학습
- 분류
- 회귀
- 예측

2. 비지도 학습
- 군집화(Clustering)
- 차원 축소(Dimension Reduction)

3. 준지도 학습

### 강화 학습
---
기본 용어
- 에이전트(Agent): 학습 주체 (혹은 actor, controller)
- 환경(Environment): 에이전트에게 주어진 환경, 상황, 조건
- 행동(Action): 환경으로부터 주어진 정보를 바탕으로 에이전트가 판단한 행동
- 보상(Reward): 행동에 대한 보상을 머신러닝 엔지니어가 설계

대표 종류
- Monte Carlo methods
- Q-Learning
- Policy Gradient methods

커뮤니티
- reinforcement-learning-kr(https://github.com/reinforcement-learning-kr)
- Awesome Reinforcement Learning(https://github.com/aikorea/awesome-rl)

### 다양한 머신러닝 알고리즘
---
1. 선형 회귀와 로지스틱 회귀


2. 선형 SVM과 커널 SVM


3. 의사결정 나무와 앙상블 트리
- 랜덤 포레스트 : 여러 개의 의사결정나무를 앙상블해 과적합 방지
- 그래디언트 부스팅 : 이전 트리의 오차를 학습해 성능 향상

4. 신경망과 딥러닝

5. K-평균(K-means), K-모드(K-modes), 가우시안 혼합 모델(GMM; Gaussian Mixture Model) 클러스터링
- K-평균 : 거리 기반 크러스터링 알고리즘
- K-모드 : 해밍거리 사용
- 가우시안 혼합 모델 : 데이터 포인트가 어떤 가우시안 분포를 따르는지 연속확률분포를 추정


6. DBSCAN(Density-Based Spatial Clustering of Applications with Noise)
- 지리적 데이터, 공간 데이터에 효과적

7. 계층적 군집화(Hierarchical clustering)
